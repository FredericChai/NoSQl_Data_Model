```python
import bson
import datetime
from pymongo import MongoClient
from pymongo import collection
import timeit
%load_ext line_profiler
%load_ext memory_profiler
```

```python
db = MongoClient('localhost', 27017)['Assignment'] #connect to mongoDB using hostname,port and database name
collection_list = db. list_collection_names() #list all collection name
print(collection_list)
```

```python
def getCollection(collectionName = 'Post'):
    collection = db[collectionName]
    return collection
```

```python
'''
SQ1
----------
index:
        Post collection: Id, ParentId,OwnerUserId
        Users collection: _id
'''
def relatedUser(Id):
    id = str(Id)
    question = getCollection(collectionName = 'Post')
    question_User_Id = question.find_one({"Id":id},{"OwnerUserId":1})['OwnerUserId']
    pipeline = [
        {"$match":{"$or":[{"ParentId":id},  #find all answers according to the given question id
                                  {"Id":id}]}},     #find all users ask the question
        {"$group":{"_id":"$OwnerUserId"}}
    ]
    all_user =[]
    for i in question.aggregate(pipeline):   #use list to load user id
        all_user.append(i['_id'])    
    user = getCollection(collectionName = 'Users') 
    pipeline_u = [
        {"$match":{"_id":{"$in":all_user}}} # use document reference of user collection
    ]
    print("the question: ",id, " and its related user is :")
    for i in user.aggregate(pipeline_u):
        print(i)
        
%time relatedUser(1)
```

```python
'''
SQ2
----------
index:
        Post collection: Id, ParentId
        Users collection: _id
'''
def SQ2(topic):
    collection = getCollection(collectionName='Topics')
    question_Id = collection.find_one({"_id":topic})['question']   #get all question id according the given tag
    pipeline = [
        {"$match":{"Id":{"$in":question_Id},"PostTypeId":"1"}},  #find all the question
        {"$sort":{"ViewCount":-1}},
        {"$limit":1},
        {"$project":{"_id":"$Id","ViewCount":1,"Title":1,"Tags":1}}
    ]
    post_question = getCollection(collectionName='Post')
    for i in post_question.aggregate(pipeline):
        print(i)

%time SQ2("deep-network") #given tag deep-network
```

```python
'''
AQ1
----------
index:
        Post collection: Id, ParentId
'''

#find the shortest answer time for each topics
def shortestAnswerTime(questionId):
#     print(questionId,':')
    post_question = getCollection('Post')
    pipeline = [
        {"$match":{"Id":{"$in":questionId},"PostTypeId":"1"}},
        {"$lookup":{"from":"Post","localField":"Id","foreignField":"ParentId","as":"Answer"}},
        {"$project":{"_id":"$Id","CreationDate":1,"Title":1,"AnswerTime":"$Answer.CreationDate"}},
        {"$unwind":"$AnswerTime"},
        {"$project":{"_id":"$_id","Title":1,
                    "Shortest_Ans_Minutes":{"$divide":[{"$subtract":["$AnswerTime","$CreationDate"]},60000]}
        }},
        {"$sort":{"Shortest_Ans_Minutes":1}},
        {"$limit":1}
    ]
    for i in post_question.aggregate(pipeline):
        print(i)
        
#output easiest for the given list       
def getEasiestAnswer(topic):
    for i in topic:
        topics = i
        collection_Topics = getCollection('Topics') 
        question_Id_list = collection_Topics.find_one({'_id':topics})['question'] #find all easiest answer for given topic
        print('Easiest question for  \"',topics,"\"")
        shortestAnswerTime(question_Id_list) 

#Main    
list_of_topics = ['value-iteration','dimensionality','discount-factor','autoencoders']
%time getEasiestAnswer(list_of_topics)
```

```python
'''
AQ2
--------
Index:
        Post collection: CreationDate
'''
def MostPopularTopic(TimeRange):
    Time = TimeRange
    answer = getCollection('Post')
    pipeline = [
        {"$match":{"CreationDate":{"$gte":TimeRange['start'],"$lte":TimeRange['end']}}},
        {"$unwind":"$Tags"},
        {"$group":{"_id":"$Tags","engage_user":{"$addToSet":"$OwnerUserId"}}},
        {"$project":{"_id":1,"NumOfUser":{"$size":"$engage_user"}}},
        {"$sort":{"NumOfUser":-1}},
        {"$limit":5}
    ]
    for i in answer.aggregate(pipeline):
        print(i)
        
TimeRange = {'start':datetime.datetime(2018, 8, 1, 0, 0, 0, 0),'end':datetime.datetime(2018, 8, 31, 0, 0, 0, 0)} #specify the time
%time MostPopularTopic(TimeRange)
```

```python
'''
AQ4
-----
index
        Post: AcceptedAnswerId_PostTypeId
'''
def getFavouriteQuestion(user = "4398", threshhold = 4):
    collection  = getCollection('Post')
    pipeline=[
        {"$match":{"AcceptedAnswerId":{"$ne":''},"PostTypeId":"1"}},
        {"$lookup":{"from":"Post","localField":"AcceptedAnswerId","foreignField":"Id","as":"Anwer"}}, #find the owner of the accepted answer
        {"$project":{"_id":1,"Tags":1,"accepted_user":"$Anwer.OwnerUserId"}}, 
        {"$match":{"accepted_user":user}},
        {"$unwind":"$accepted_user"},
        {"$unwind":"$Tags"},
        {"$group":{"_id":"$Tags","Number_of_AccpetedAnswer":{"$sum":1}}}, # count the number of accepted answer
        {"$match":{"Number_of_AccpetedAnswer":{"$gte":threshhold}}},
        {"$sort":{"Number_of_AccpetedAnswer":-1}},
        {"$limit":5}   #output the first five label
         ]
    result = []
    label = []
    for i in collection.aggregate(pipeline):
        result.append(i)
    for i in result:
        label.append(i['_id'])
    print("The question that user answered most is: ",label)
    collection = getCollection('Post')
    pipeline = [
        {"$match":{"AcceptedAnswerId":'',"PostTypeId":'1'}},
        {"$project":{"_id":"$Id","Title":1,"Tags":1,"date":"$CreationDate"}},
        {"$unwind":"$Tags"},
        {"$match":{"Tags":{"$in":label}}},  #read the label statify the conditions 
        {"$group":{"_id":"$_id","Title":{"$first":"$Title"},"date":{"$first":"$date"},"Tags":{"$addToSet":"$Tags"}}},  # group the questions
        {"$match":{"date":{"$lte":datetime.datetime(2018, 8, 30, 0, 0, 0, 0)}}},
        {"$sort":{"date":-1}},
        {"$limit":5}
    ]
    print("The recommended question is :")
    for i in collection.aggregate(pipeline):
        print(i)

%time getFavouriteQuestion()
```

 [AQ3] Given a topic, nd the champion user and all questions the user has answers accepted in that topic.

We dene a champion user as the one who has the most answers accepted in a given topic. For instance, the champion user of topic deep-learning is 4398 and 1847. Both users have 9 answers being accepted in this topic. Your result may show either of the two users. Below is an example of the questions user 4398 has answers accepted:


```python
'''
AQ3
--------
Index:
        Post collection: Id
'''
def ChampionAnswer(Topic):
    topic = "deep-learning"
    collection = getCollection('Topics')
    pipeline = [
        {"$match":{"_id":topic}},
        {"$unwind":{"path":"$question","preserveNullAndEmptyArrays":True}},
        {"$lookup":{"from":"Post","localField":"question","foreignField":"Id","as":"Q"}},
        {"$project":{"_id":"$question","champion_answer":"$Q.AcceptedAnswerId"}},
        {"$unwind":"$champion_answer"},
        {"$match":{"champion_answer":{"$ne":''}}}, #throw documents where champion_answer are null
        {"$lookup":{"from":"Post","localField":"champion_answer","foreignField":"Id","as":"Answer"}},
        {"$project":{"_id":1,"champion_user":"$Answer.OwnerUserId"}},
        {"$unwind":"$champion_user"},
        {"$group":{"_id":"$champion_user","questionId":{"$addToSet":"$_id"},"AccpetTimes":{"$sum":1}}},
        {"$sort":{"AccpetTimes":-1}},
        {"$limit":2} 
    ]
    result = []
    for i in collection.aggregate(pipeline):
        result.append(i)       #get the champion answer and append their detail in the result
    collection_question = getCollection('Post')
    for i in range(len(result)):
        print("The champion answer user ", result[i]['_id'], "and the corresponding question is:")
        pipeline = [
            {"$match":{"Id":{"$in":result[i]['questionId']}}},
            {"$project":{"Id":1,"Title":1,"_id":0}},
            {"$sort":{"Id":1}}
        ]
        for i in collection_question.aggregate(pipeline):
            print(i)

topic = 'deep-learing'
%time ChampionAnswer(topic)
```

```python
'''
AQ6
--------
This query are used to compare with the neo4j which are not be included in the mongoDB work load

'''
def findCoAuthor(id):
    user_id = str(id)
    collection = getCollection("Post")
    pipeline = [
        {"$match":{"OwnerUserId":user_id}},   #given user
        {"$project":{"_id":"$Id","PostTypeId":1,"OwnerUserId":1,"ParentId":1}},
         {"$lookup":{         # embedd the questioner whose question are answered by this given user
                    "from":"Post",
                    "localField":"ParentId",
                    "foreignField":"Id",
                    "as":"HelpedQuestioner"
        }},
         {"$graphLookup":{     # embedd the answerer who answer the same question with this given user
                    "from":"Post",
                    "startWith":"$ParentId",
                    "connectFromField":"Id",
                    "connectToField":"ParentId",
                    "as":"SameLabelAnswer"
        }},
        {"$graphLookup":{         # embedd the answerer who answer the question this user asked
                    "from":"Post",
                    "startWith":"$_id",
                    "connectFromField":"_id",
                    "connectToField":"ParentId",
                    "as":"CooperatedAnswer"
        }},
         {"$project":{"_id":"$Id","PostTypeId":1,"OwnerUserId":1,"ParentId":1,
                      "CooperatedAuthor":"$SameLabelAnswer.OwnerUserId",
                          "cooperatedAnswerer":"$CooperatedAnswer.OwnerUserId",
                          "helpedQuestioner":"$HelpedQuestioner.OwnerUserId"
                          }},
         {"$project":{"_id":"$Id","PostTypeId":1,"OwnerUserId":1,"ParentId":1,
                      "CooperatedAuthor":{"$setUnion":["$CooperatedAuthor",
                                                  "$cooperatedAnswerer",
                                                  "$helpedQuestioner"
                                                    ]}
                     }},
        {"$unwind":{"path":"$CooperatedAuthor","preserveNullAndEmptyArrays":True}},
        {"$group":{"_id":"$CooperatedAuthor","CoopTime":{"$sum":1}}},
        {"$match":{"_id":{"$ne":user_id}}},
        {"$match":{"_id":{"$ne":None}}},
        {"$sort":{"CoopTime":-1}},
        {"$limit":5}
    ]
    for i in collection.aggregate(pipeline):
        print(i)

%time findCoAuthor('4398')
```

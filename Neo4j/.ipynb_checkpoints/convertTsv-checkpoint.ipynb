```python
from py2neo import *
import csv
```

```python
#preprocess post
# Id	PostTypeId	AcceptedAnswerId	CreationDate	
# Score	ViewCount	OwnerUserId	Title	Tags	AnswerCount	CommentCount	
bucket= []
with open("Posts.tsv") as post:
        read_post = csv.reader(post, delimiter='\t',quotechar='"')
        next(read_post)
        for line in read_post:
            temp = [
                    line[0], #'Id'
                    line[1], #'PostTypeId'
                    line[2], # 'AcceptedAnswerId'
                    line[3],#'CreationDate'
                    line[5], #'ViewCount':
                    line[6], #'OwnerUserId'
                    line[7], #'Title'
                    line[8],#'Tags'
                    line[12]#'ParentId'
            ]    
            bucket.append(temp)
        
with open('post.csv','w') as post:
        write_post = csv.writer(post,delimiter=',',quotechar='"')
        header = ['Id','PostTypeId','AcceptedAnswerId','CreationDate','ViewCount','OwnerUserId','Title','Tags','ParentId']
        write_post.writerow(header)
        write_post.writerows(bucket)
```

```python
#preprocess votes
bucket = []
with open("Votes.tsv") as vote:
        read_vote = csv.reader(vote, delimiter='\t',quotechar='"')
        next(read_vote)
        for row in read_vote:
            temp = [
                row[0], #Id': 
                row[1], # 'PostId':
                row[2], # 'VoteTypeId': 
#                 'CreationDate': row[3],
#               'UserId': row[4], 
#                 'BountyAmount': row[5]
            ]
            bucket.append(temp)
        
with open('/Users/chaizhizhi/Desktop/5338Assignment/Neo4jCypher/vote.csv','w') as post:
        write_post = csv.writer(post,delimiter=',',quotechar='"')
        header = ['Id','PostId','VoteTypeId']
        write_post.writerow(header)
        write_post.writerows(bucket)
```

```python
#preprocess user
bucket = []
with open("Users.tsv") as User:
        read_User = csv.reader(User, delimiter='\t',quotechar='"')
        next(read_User)
        for row in read_User:
            temp = [
              row[0], #  'Id': 
#                     'Reputation': row[1], 
                    row[2], #'CreationDate': 
                     row[3], #'DisplayName':
#                     'LastAccessDate': row[4], 
#                     'Location': row[5], 
#                     'Views': row[6], 
                  row[7],#  'UpVotes': 
                    row[8],  #'DownVotes': 
#                 'AccountId': row[9]
            ]
            bucket.append(temp)
        
with open('/user.csv','w') as post:
        write_post = csv.writer(post,delimiter=',',quotechar='"')
        header =  ['Id','CreationDate','DisplayName','UpVotes','DownVotes']
        write_post.writerow(header)
        write_post.writerows(bucket)
```

```python
#preprocess tags
bucket = []
with open("Tags.tsv") as tag:
        read_tag = csv.reader(tag, delimiter='\t',quotechar='"')
        next(read_tag)
        for row in read_tag:
            temp = [
#               row[0],  #'Id': 
                   row[1] # 'Tagname': 
#                     'Count': row[2],
#                     'ExcerptPostId': row[3], 
#                     'WikiPostId': row[4] 
            ]
            bucket.append(temp)
        
with open('tag.csv','w') as post:
        write_post = csv.writer(post,delimiter=',',quotechar='"')
        header =  ['topic']
        write_post.writerow(header)
        write_post.writerows(bucket)
```
